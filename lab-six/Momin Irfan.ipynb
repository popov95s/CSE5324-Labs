## Business Understanding

### Overview

We will be using a collection of images of dogs, sorted by breed, as our dataset for this lab.  The dataset can be found at [footnote 1](#footnote1); it was created by Stanford graduate students and their advisor as a data set specifically for fine-grained image categorization. Besides the categorized pictures of dogs, the data set also includes bounding box annotations for each of the images, class labels, and information on how the creators used the set to train and test.  The full set contains 20,560 images, with 120 different breeds of dogs represented.  There are at least 148 images of each breed, with the Maltese dog having the most images at 252.  Each picture is at least 200 pixels by 200 pixels in size, with most pictures generally not being much larger than that.

### Purpose

This data is a specifically created set of images, created using images found on ImageNet under the 'Canis familiaris' node.  The three graduate students, Khosla, Jayadevaprakash, and Yao, along with their mentor, Fei-Fei, created this set of images specifically with the purpose of creating a dataset for fine-grained image categorization.  They went through the images found on ImageNet one breed at a time, being sure to select images that were only classified under one node, and removed images which had distortion, major occlusion, and other defects, in order to create a subset of images that they believed addressed the drawbacks of other fine-grained image categorization datasets.  Further information about their process and their results in achieving their dataset goals can be found at [footnote 2](#footnote2).

### Prediction Task

The prediction task for this dataset is relatively simple: to be able to recognize the breed of a dog from a picture, which may be of different age, color, shape than the original images, or which may be partially occluded. Generally, there is little inter-class variation in the appearance of dogs <sup>[[1]](#footnote1)</sup> and large intra-class variation<sup>[[1]](#footnote1)</sup>, making recognizing dogs a tough challenge for categorization.  Even with a human, there are certain types of dogs that pose challenges when it comes to differentiation, especially among mixed breed dogs; to be able to identify the type of dog in an image mechanically would help in cases as simple as a person trying to identify what breeds are in a mongrel's ancestry. <br><br>
At a more advanced level, it would be helpful to many photo sharing websites to be able to recognize a type of dog from the image posted, in order to create more tailored advertising for said person.  To be able to recognize a dog is to be able to advertise generalized dog products at a user, while to be able to recognize breed is to be able to target a user for very specific dog-related purchases, which may drive higher advertising revenue.  It has been demonstrated how much more effective targeted ads can be than their non-targeted counterparts<sup>[[3]](#footnote3)</sup>; to continue to truly personalize ads based on the very specifics of a person's lifestyle is the ultimate goal of web advertisment. Advertising is what drives the modern web, and the modern dog owner spends \$1549<sup>[[4]](#footnote4)</sup> per year on dog-related purchases, to be able to further target advertising has the potential to be very valuable. There is likely no specific level of accuracy required for such use, as long as the classifier is somewhat accurate.  Generally, customers simply ignore ads that do not relate to them<sup>[[11]](#footnote11)</sup>, so while there is an opportunity cost to not having a perfectly accurate classifier, there is likely to be no material cost.  Even with just the ability to tell between a small dog and a big dog, advertising can be more focused; the increase in accuracy is nonessential, but useful nonetheless. <br><br>
Besides helping those mentioned above, the ability to identify dogs is one that has major implications to veterinarians and others in industries working directly with dogs.  When it comes to caring for dogs, there are material differences between certain breeds. Different breeds vary in their needs, including requiring different nutrition<sup>[[5]](#footnote5)</sup> and different levels of exercise<sup>[[6]](#footnote6)</sup>; most important, however, is dosage of veterinary drugs. There are certain drugs used by vets that are important to the health of a dog, but have the potential to be deadly to certain breeds which carry specific gene mutations<sup>[[7]](#footnote7)</sup>.  To be able to identify a dog, and avoid using a drug that could have fatal side effects is invaluable.  With a mechanical method of dog identification, vets can be more certain that they are treating a dog the correct way for its genetics, and are advising the owners of the correct way to feed and take care of it on a regular basis. Just like doctors in other fields, it is unlikely that veterinary physicians would use a computer as an infallible resource, rather they would use it to assist and support in the diagnostics they already practice<sup>[[12]](#footnote12)</sup>.  Due to such usage, where it is unlikely that egregious errors will ever make it through the process, the level of accuracy required would depend on the amount of time the classifier takes.  If the vet were to spend a significant amount of time using the classifier, and consistently got results that were incorrect, it is unlikely that the classifier would see any use.  Ideally the classifier would hit 90% accuracy, the level of accuracy found in DNA dog tests<sup>[[13]](#footnote13)</sup>, but users would likely accept lower accuracy in general for reasons of convenience.<br><br>
Another field in which being able to identify a dog is important is pet adoption and animal shelters. One in ten dogs adopted from shelters in the United States are returned within six months, most often due to unexpected costs or behaviors<sup>[[8]](#footnote8)</sup>.  Many adopt dogs not knowing the dog's exact genetic makeup, and are unaware of what the dog will grow into from the puppy they adopt.  While it is impossible to fully predict a dog's size and behavior simply from its ancestry, the ancestry is still one of the best indicators of what a dog will be like as an adult.  To be able to tell what kind of dog a puppy will become can assist in the operations of pet shelters and other adoption organiztions, giving them the ability to inform the customer what they can expect in the future from any given dog. When it comes to the accuracy of shelters identifying dog breeds, most are not very reliable.  Only 12.5% of dogs identified by shelters match with the shelter-described breeds exactly<sup>[[14]](#footnote14)</sup>, so any accuracy above 12.% is an improvement. However, one must take into account how egregious the errors of the shelters were; the degree of error may matter more than the incidence of error. <br><br>
Finally, it can be important to identify dogs in order to know that one is receiving the actual type of dog that they are told they are receiving.  Only two states currently offer statutory recourse for the misrepresentation of a breed in purchasing a dog: Florida and Minnesota<sup>[[9]](#footnote9)</sup>. In other states, it is possible to attempt recourse through various other methods, such as breach of contract, but it is not always guaranteed that the customer can be assured that they will receive exactly what they wanted.  With a method of identifying dogs that does not rely on human trust, customers can feel more confident in their purchases of dogs, knowing that they are almost certainly purchasing the correct merchandise.  The accuracy of a classifier is not as vital in such cases, due to the rarity of the situation, and the avenues of recourse already in place.  If a customer is already distrustful enough of a business to use such a system, it is likely that they are simply looking for something to assist them in confirming their suspicions; at the point of purchase, not much is on the line for either side of the transaction.  Again, if the classifier could outdo DNA classification, and improve over 90% accuracy, it would be ideal, and perhaps become the new standard of verifying pedigree, but for the most part in such scenarios it would likely serve as a method to help the customer overcome any cognative dissonance.

### Third Parties

The third parties that would be interested in such a classification system would be a combination of social media sites and advertisers.  Working in unison, they would be able to look at the pictures posted on the social media account of a user, and not only determine what kind of pet they have, they could go further and determine the subtype of that pet, in order to maximize personalization of the ads.  The most obvious example of a site that would use a system such as this one would be Instagram. Instagram is entirely based around lifestyle and images, and pet photos are very common. Overall, any site that relies on user-generated content and advertising would likely reap the benefits of being able to classify pets as granularly as possible. <br><br>
Besides the third parties in technology, anyone in the fields of veterinary medicine and dog breeding, or even just those that love dogs, could use the app to help in the pursuit of those things.  Whether it is ensuring that the medicine being given is appropriate, confirming that a dog is what it is being said, or just trying to identify a cute dog seen in one's neighborhood, the ability to identify or confirm the human identification of a dog is a valuable resource.  In the United States, 36.5% of households own dogs<sup>[[10]](#footnote10)</sup>; it is the largest pet market in the country.  From owning dogs, to selling dogs, to treating dogs, the identification of dogs runs through American life; lending a helping hand through technology could touch millions of lives, helping both the humans and their dogs.

## Evaluation Criteria: 

We will use accuracy to evalueate our model. Although accuracy is not usually the best metric, accuracy will suffice in giving us an accurate depiction of how our specific model performs. Our team was exploring whether or not to focus on Sensitivity or Recall, but decided that given the charachteristics of our dataset and our classification problem, accuracy would be a better measure.

The reasons for this selection are as follows: 

### Classification Problem:
Our model aims to loabel dog images with the breed of dog in the photo. It is an incredibly large multi-class problem and a fairly complex task. The use cases for this model stem from the areas of pet health, pet adoption/sales, and pet image labeling. Of these use cases, pet health has highest cost for misclassification, then adoption/sales, and then image labeling. 

With the health use case, we would like to have both recall and precision. Recall in our case (using corgis as an example) means that we can identifiy as many corgis correctly. Precision means that when we label something as a corgi, there is a good chance to beleive that this image is a corgi. The health case needs both precision and recall, because the ways we can fail (not giving a dog the proper care) is either mistreating a dog (confusing breed) or not recognizing the breed of the dog. To have both of these, we must have both percision and recall. 

The dog sale/adoption use case is very similar. The biggest cost in this case is posting an incorrect "advertisement". This can come with legal risks and business risks. So in this case what is most important is precision. There is a case to be made that recall is also very important (opportunity costs), but those do not become customer facing issues so recall is not as important as precision in this case. 

With the final use case, object recognition, the dogs will generally be labeled in photos and the like. Mislabeling the breed does not have a large cost, as these systems will not be in crucial locations, so generall recall/precision will be important. 

Given all of these classification problems, it is apparant that both recall and precision are important. Thus we would use a metric that incorporates both, and that would be accuracy. 

### Dataset:

Given that we would like to use accuracy, we have to first make sure our dataset is balanced class wise. Given that there are over 20,000 images and the largest class has around 200 images, there is no large dominant class. This means that accuracy will still be a very informative metric in our tests. 

### Evaluation Metrics: 

We will use accuracy as our main evaluation metric. We will explore precision and recall when need be, although we beleive accuracy does well enough to contain both of those metrics. 


## Train Test Split:

We will use a 10-fold stratified cross validation for our training and testing. We chose this cross-validation method given the nature of our data set and how our model will be used in practice. The 10-fold aspect of this method allows us to test the performance amongst many test sets so we get an aaccurate sense of generalization performance. 

### Data Set:
A defining charachteristic of our dataset is the large multi-class aspect of it. There are many labels to predict with no real dominant dataset (200 max out of 20,000). This led us to want to stratify our set, so we are able to train on every class. There are generally the same amount of photos by class so we would like to keep that relationship. 

### Usage: 
The usage of this model in practice will be trained over a large corpus of dogs. This is because in practice no body knows which breeds will be predicted, there is a wide range of possible breeds. Thus, in practice the users will create a large set that they will use to train the model. This set can grow as more breeds are added into the dataset. 

Furthermore, There is no distinction between old photos or new photos, just any photos of that breed. Once a set can recognize a breed well, the images of that set will be solidified. A stratified strategy mimics the idea that we are training on a huge corpus (same ratio of pictures) and allows for a large hollistic training. It keeps the ratios the same accross the tests to keep is standard for our generalization testing.  

### Conclusion: 
Thus, we will use 10-fold cross validation to test our data.

```python
#adapted from http://dandxy89.github.io/ImageModels/alexnet/
DROPOUT = 0.5
input_layer = Input(shape = (75, 100, 3))
# First convolutional Layer (96x11x11)
x = Conv2D(filters = 96, kernel_size = (11,11), strides = (4,4), activation = "relu")(input_layer)
x = MaxPooling2D(pool_size = (3,3), strides=(2,2), data_format='channels_last')(x)
x = BatchNormalization()(x)

# Second convolutional Layer (256x5x5)
x = ZeroPadding2D(padding = (2,2))(x)
x = Convolution2D(filters = 256, kernel_size = (5,5), strides = (1,1), activation = "relu")(x)
x = MaxPooling2D(pool_size = (3,3), strides=(2,2))(x)
x = BatchNormalization()(x)

# Rest 3 convolutional layers
x = ZeroPadding2D(padding = (1,1))(x)
x = Convolution2D(filters = 384, kernel_size = (3,3), strides = (1,1), activation = "relu")(x)

x = ZeroPadding2D(padding = (1,1))(x)
x = Convolution2D(filters = 384, kernel_size = (3,3), strides = (1,1), activation = "relu")(x)

x = ZeroPadding2D(padding = (1,1))(x)
x = Convolution2D(filters = 256, kernel_size = (3,3), strides = (1,1), activation = "relu")(x)

x = MaxPooling2D(pool_size = (3,3), strides=(2,2))(x)
x = Flatten()(x)

x = Dense(4096, activation="relu")(x)
x = Dropout(DROPOUT)(x)

x = Dense(4096, activation="relu")(x)
x = Dropout(DROPOUT)(x)

final_dim = NUM_CLASSES
final_act = "softmax"
output_layer = Dense(final_dim, activation=final_act)(x)

model = Model(input_layer, output_layer)
model.summary()
```

## ResNet34

In this section, we train the 34 layer ResNet architecture mentioned above.

```python
from keras.layers import average, concatenate, Add
from keras.models import Input, Model
from keras.regularizers import l2
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D

resnet_layers = [(64,False),(64,False),(64,False),
                 (128,True),(128,False),(128,False),(128,False),
                (256,True),(256,False),(256,False),(256,False),(256,False),(256,False),
                (512,True),(512,False),(512,False)]
l2_lambda = .000001
sh = (100,75,3)

inputs = Input(shape=sh)

x = Conv2D(filters=64,
              input_shape=sh,
              kernel_size=(7,7),
              strides=(2,2),
              kernel_initializer='he_uniform',
              kernel_regularizer=l2(l2_lambda),
              padding='same',
              activation='relu',
              data_format='channels_last')(inputs)
x = MaxPooling2D(pool_size=(2,2),strides=(2,2), data_format='channels_last')(x)

for num_filters, stride in resnet_layers:
    x_split = x
    if stride:
        x = Conv2D(filters=num_filters,
                  input_shape=sh,
                  kernel_size=(3,3),
                  strides=(2,2),
                  kernel_initializer='he_uniform',
                  kernel_regularizer=l2(l2_lambda),
                  padding='same',
                  activation='relu',
                  data_format='channels_last')(x_split)
    else:
        x = Conv2D(filters=num_filters,
                  input_shape=sh,
                  kernel_size=(3,3),
                  kernel_initializer='he_uniform',
                  kernel_regularizer=l2(l2_lambda),
                  padding='same',
                  activation='relu',
                  data_format='channels_last')(x_split)
    x = Conv2D(filters=num_filters,
              input_shape=sh,
              kernel_size=(3,3),
              kernel_initializer='he_uniform',
              kernel_regularizer=l2(l2_lambda),
              padding='same',
              activation='relu',
              data_format='channels_last')(x)

    if stride:
        x_split = Conv2D(filters=num_filters,
                  input_shape=sh,
                  kernel_size=(1,1),
                  strides=(2,2),
                  kernel_initializer='he_uniform',
                  kernel_regularizer=l2(l2_lambda),
                  padding='same',
                  activation='relu',
                  data_format='channels_last')(x_split)
    x = Add()([x,x_split])
    x = Activation('relu')(x)
    #if pooling:
        #x = MaxPooling2D(pool_size=(2,2), data_format='channels_last')(x)
x = AveragePooling2D(pool_size=(2,2), data_format='channels_last')(x)

#Add more layers

x = Flatten()(x)
x = Dropout(0.25)(x)
x = Dense(1000)(x)
x = Activation('relu')(x)
x = Dropout(0.5)(x)
x = Dense(NUM_CLASSES)(x)
x = Activation('softmax')(x)

resnet34 = Model(inputs=inputs,outputs=x)

resnet34.summary()
```

```python
%%time

from keras.callbacks import EarlyStopping
# Let's train the model 

resnet34.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

history_res34 = resnet34.fit_generator(datagen.flow(X_train_reshaped,y_train_ohe,batch_size=32),
                    steps_per_epoch=int(len(X_train_reshaped)/32),
                    epochs=20,verbose=1,
                    validation_data=(X_test_reshaped,y_test_ohe),
                    callbacks=[EarlyStopping(monitor='val_loss',patience=5)])

resnet34.save_weights('final_weights_res_34.h5')
```

# Exceptional Work

For exceptional work, we use transfer learning. We specifically use the implementation of ResNet50 from Keras' applications. This is a 50 layer resnet that has been pretrained on ImageNet data. Since ResNet50 was trained on images that are 244 by 244 pixels, we must resize our images to fit this size. Additionally, ResNet50 was trained on color images, so we preserve the color in the images rather than converting them to grayscale.

We use ResNet50 as the first layers of our network, appening onto it a Dense layer (1024 nodes) and an output Dense layer that will output which dog breed is represented by the input image.

```python
img_size = (224,224,3)
img_path = '../Images/'
img_paths = []
img_annotations = []
N_IMAGES = 4000

#The images are stored in subdirectories - go through all of them
for dirpath, dirnames, filenames in os.walk(img_path):
    for filename in [f for f in filenames if f.endswith(".jpg")]:
        img_paths.append(os.path.join(dirpath, filename))
        img_annotations.append(dirpath.split('/')[-1].split('-')[1])
img_annotations = img_annotations[0:N_IMAGES]

img_data=[]
for i,path in enumerate(img_paths[0:N_IMAGES]):
    img_data.append((resize(plt.imread(path).astype(np.uint8),img_size)).flatten())

X = np.array(img_data)
unique_classes = list(set(img_annotations))
class_dict = {unique_classes[x]:x for x in range(len(unique_classes))}
y = np.array([class_dict[i] for i in img_annotations])
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import OneHotEncoder
from copy import deepcopy

skf = StratifiedKFold(n_splits = 10)

train_list = []
test_list = []

for train_index, test_index in skf.split(X, y):
    train_list.append(train_index)
    test_list.append(test_index)

X_train_transfer = deepcopy(X[train_list[0]])
X_test_transfer = deepcopy(X[test_list[0]])
y_train_transfer = deepcopy(y[train_list[0]])
y_test_nohe = deepcopy(y[test_list[0]])

NUM_CLASSES = len(unique_classes)

input_size= img_size[0]*img_size[1] if len(img_size)==2 else img_size[0]*img_size[1]*img_size[2]
```

```python
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np

res_model = ResNet50(weights='imagenet')

x = res_model.output
x = Dense(1024, activation='relu',kernel_initializer='he_uniform')(x)
x = Dense(NUM_CLASSES,activation='softmax',kernel_initializer='glorot_uniform')(x)
model = Model(inputs = res_model.input, outputs=x)
for layer in res_model.layers:
    layer.trainable = False

model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['categorical_accuracy'])
```

```python
y_train_transfer = keras.utils.to_categorical(y_train_transfer, NUM_CLASSES)
y_test_transfer = keras.utils.to_categorical(y_test_nohe, NUM_CLASSES)
X_train_transfer = np.reshape(X_train_transfer,(X_train_transfer.shape[0],*img_size))
X_test_transfer = np.reshape(X_test_transfer,(X_test_transfer.shape[0],*img_size))
print(X_train_transfer.shape)
```
